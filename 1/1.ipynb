```python
import numpy as np
```

```python
#Koordinate poznatih 7 temena (na osnovu Paint-a)
p1 = np.array([493,495])
p2 = np.array([270,687])
p3 = np.array([129,553])
p5 = np.array([524,357])
p6 = np.array([268,524])
p7 = np.array([106,405])
p8 = np.array([359,294])
```

```python
def affToHom(point):
    return np.append(point,1)
```

```python
def homToAff(point):
    return np.round(point[:2])
```

```python
p1
```

```python
p1H = affToHom(p1)
```

```python
p2H = affToHom(p2)
p3H = affToHom(p3)
p5H = affToHom(p5)
p6H = affToHom(p6)
p7H = affToHom(p7)
p8H = affToHom(p8)
```

```python
def afinize(point):
    return point/point[2] 
```

```python
#Funkcija koja racuna presek dve prave i vraca tacku preseka u obliku (x,y,1)
def intersect(point1, point2, point3, point4):
    l1 = np.cross(point1,point2)
    l2 = np.cross(point3, point4)
    return afinize(np.cross(l1,l2))
```

```python
# Funkcija koja za odredjuje tacke nedogleda na sledeci nacin:
# Tacka nedogleda je presek 'paralelnih' pravih. Na nasem kvadru mi imamo tri para takvih pravih po jednom odredjenom pravcu, 
# Zbog cega racunamo za svaki par pretpostavljenu tacku nedogleda
# I konacan rezultat racunamo kao njihov prosek
def vanishingPoint(point1, point2, point3, point4, point5, point6):
    x1 = intersect(point1,point2,point3,point4)
    x2 = intersect(point1, point2, point5, point6)
    x3 = intersect(point3, point4, point5, point6)
    return (x1 + x2 + x3)/3
```

```python
X = vanishingPoint(p1H, p2H, p5H, p6H,p7H, p8H)
Y = vanishingPoint(p2H, p3H, p6H, p7H, p5H, p8H)
Z = vanishingPoint(p3H, p7H, p2H, p6H, p1H, p5H)

#Nedostajuce teme p4 racunamo kao presek prava iz tacaka nedogleda sa vec poznatim odgovarajucim temenima
#To radimo na isti nacin na koji racunamo i tacke nedogleda
#Izracunamo svaki od njih i opet uzimamo njihov prosek kao konacnu tacku
p4 = vanishingPoint(X, p3H, Z, p8H, Y, p1H)
homToAff(p4)
```
